```python
df = pd.read_csv("/kaggle/input/linear-regression-data-csv/linear_regression_dataset.csv", sep=";")
```

```python
df.head()
```

```python
plt.scatter(df.deneyim,df.maas) 
plt.xlabel("Deneyim") 
plt.ylabel("Maas")
plt.show()
```

```python
x = df.deneyim.values
```

```python
x.shape
```

```python
x = df.deneyim.values.reshape(-1,1)
y = df.maas.values.reshape(-1,1)
```

```python
from sklearn.linear_model import LinearRegression
linear_reg = LinearRegression()
linear_reg.fit(x,y)
```

```python
import numpy as np
b0 = linear_reg.predict([[0]])
print("b0: ",b0)
```

```python
b0_ = linear_reg.intercept_
print("b0: ",b0_)
```

```python
b1 = linear_reg.coef_
print("b1: ",b1)
```

```python
new_salary = 1663 + 1138*11
print(new_salary)
```

```python
b11 = linear_reg.predict([[11]])
print("b11: ",b11)
```

```python
y_head = linear_reg.predict(x)
```

```python
plt.plot(x, y_head, color="green")
plt.scatter(x,y)
plt.show()
```

```python
from sklearn.metrics import r2_score
print("R Square Score: ",r2_score(y,y_head))
```

```python
from sklearn.metrics import mean_squared_error
print("Mean Squared Error: ",mean_squared_error(y,y_head))
```

```python
MSE = np.square(np.subtract(y,y_head)).mean()
print("Mean Squared Error: ",MSE)
```
